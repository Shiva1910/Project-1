```python
q = data_1['serum_creatinine'].quantile(0.99)
data_1 = data_1[data_1['serum_creatinine']<q]
data_1.describe(include='all')
```

```python
X=data_1.drop('DEATH_EVENT',axis=1)
y=data_1['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions =logmodel.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
tit = pd.DataFrame({'Original_Survival':y_test,'Predicted':predictions})
tit
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
Pred_sur = random_forest.predict(X_train)
random_forest.score(X_train, y_train)
acc_random_forest = round(random_forest.score(X_train, y_train) * 100, 2)
acc_random_forest
```

```python
result = pd.DataFrame({'DEATH_EVENT':y_train,'Predicted':Pred_sur,'age':X_train['age'],
                   'Sex':X_train['sex']})
result
```
